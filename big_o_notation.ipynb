```python
x = np.linspace(1,10,10)
y1=0*x+1
y2=np.log2(x)
y3=x
y4=x*np.log2(x)
y5=x**2
y6=2**x
y7=[np.math.factorial(i)for i in x]
y8=x**x
fig,ax = plt.subplots()
ax.plot(x,y1,label='$0(1)$')
ax.plot(x,y2,label='$0(\log n)$')
ax.plot(x,y3,label='$0(n)$')
ax.plot(x,y4,label='$0(n\lg n)$')
ax.plot(x,y5,label='$0(n^2)$')
ax.plot(x,y6,label='$0(2^n)$')
ax.plot(x,y7,label='$0(n!)$')
ax.plot(x,y8,label='$0(n^n)$')
ax.set_title('Big_O_Notation')
ax.set_xlabel('Elements')
ax.set_ylabel('Operations')
ax.legend()
plt.show()
```
