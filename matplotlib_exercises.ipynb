```python
# : represents a dotted line
# -- represents a dashed line
import math

x = range(-4, 4)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls=':')
plt.title('An approximation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c='orange')
plt.plot(x, z, c='green')

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5),
             arrowprops={'facecolor': 'blue'})

plt.show()
```

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')

plt.show()
```

```python
# Use matplotlib to plot the following equation:
# y=X^2-X+2
#You'll need to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the origin.
def f(x):
    return x**2 -x+2
x = np.linspace(-10,10,100)
y=f(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y=x^2-x+2')
plt.annotate('Origin (0,0)', xy=(0, 3), xytext=(0, 15),
             arrowprops={'facecolor': 'red'})
plt.show()
```

```python
# Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# y=√x
# y=x^3
# y=2^X
# y=1/(x+1)
#You can use functions from the math module to help implement some of the equations above.
import math
x = np.linspace(0,10,100)
def f1(x):
    return np.sqrt(x)
plt.subplot(2,2,1)
plt.plot(x, f1(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=sqrt(2)')
def f2(x):
    return x**3
plt.subplot(2,2,2)
plt.plot(x, f2(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^3')
def f3(x):
    return 2**x
plt.subplot(2,2,3)
plt.plot(x, f3(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=2^x')
def f4(x):
    return 1/(x+1)
plt.subplot(2,2,4)
plt.plot(x, f4(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=1/(x+1)')
plt.tight_layout
plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
x = np.linspace(0,10,100)
def f1(x):
    return np.sqrt(x)
def f2(x):
    return x**3
def f3(x):
    return 2**x
def f4(x):
    return 1/(x+1)
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,8))
ax1.plot(x,f1(x))
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('y=sqrt(x)')
ax2.plot(x,f2(x))
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('y=x^3')
ax3.plot(x,f3(x))
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('y=2^x')
ax4.plot(x,f4(x))
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_title('y=1/(x+1)')
plt.tight_layout()
plt.show()
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.
x = np.linspace(0,10,100)
y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = [1/(i+1) for i in x]
plt.figure(figsize =(8,6))
plt.plot(x,y1,color='red',label='Square Root of x')
plt.plot(x,y2,color='blue',label='x^3')
plt.plot(x,y3,color='green',label='2^x')
plt.plot(x,y4,color='orange',label='1/(x+1)')
plt.legend()
plt.show()
```
