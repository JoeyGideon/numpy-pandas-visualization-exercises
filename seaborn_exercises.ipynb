```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns
```

```python

# Load the iris dataset from Seaborn
iris = sns.load_dataset('iris')
iris.head()
```

```python
# What does the distribution of petal lengths look like?
sns.histplot(data=iris, x='petal_length')
plt.title('Distribution of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.show()
```

```python
# Is there a relationship between petal length and petal width?
sns.scatterplot(data=iris, x='petal_length', y='petal_width')
plt.title('Relationship Between Petal Length and Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length?
sns.scatterplot(data=iris, x='sepal_width', y='sepal_length', hue='species')
plt.title('Sepal Width vs. Sepal Length by Species')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Which features would be best used to predict species?
sns.pairplot(data=iris, hue='species')
plt.show()
```

Load the swiss dataset from pydataset and read it's documentation. Create visualizations to answer the following questions:


```python
swiss = data('swiss')
swiss.head()
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)
cutoff = 50
swiss['is_catholic'] = swiss['Catholic']> cutoff
print(swiss.head())
```

```python
# Does whether or not a province is Catholic influence fertility?

from scipy.stats import ttest_ind

# Calculate the average fertility rate for Catholic and non-Catholic provinces
fertility_catholic = swiss[swiss['is_catholic']]['Fertility']
fertility_non_catholic = swiss[~swiss['is_catholic']]['Fertility']
mean_fertility_catholic = fertility_catholic.mean()
mean_fertility_non_catholic = fertility_non_catholic.mean()

# Perform a two-sample t-test to compare the average fertility rates of Catholic and non-Catholic provinces
t_stat, p_value = ttest_ind(fertility_catholic, fertility_non_catholic)

# Print the results
print(f"Average fertility rate for Catholic provinces: {mean_fertility_catholic:.2f}")
print(f"Average fertility rate for non-Catholic provinces: {mean_fertility_non_catholic:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.2f}")

```

```python
# What measure correlates most strongly with fertility?
correlations = swiss.corr()['Fertility'].sort_values(ascending=False)

print(correlations)
```

```python
from env import host, username, password
url = f'mysql+pymysql://{username}:{password}@{host}/chipotle'
```

```python
# Load the chipotle dataset from SQL
# create a bar chart that shows the 4 most popular items and the revenue produced by each.

# Load the Chipotle dataset from SQL
chipotle = pd.read_sql_query("SELECT * FROM orders", url)

# Clean up the dataset by converting the 'item_price' column to a float and creating a 'revenue' column
chipotle['item_price'] = chipotle['item_price'].str.replace('$', '').astype(float)
chipotle['revenue'] = chipotle['item_price'] * chipotle['quantity']

# Group the dataset by item name and sum the revenue for each item
item_revenue = chipotle.groupby('item_name')['revenue'].sum()

# Sort the items by revenue and select the top 4
top_items = item_revenue.sort_values(ascending=False)[:4]

# Create a bar chart of the top 4 items and their revenue
plt.bar(top_items.index, top_items.values)
plt.title('Top 4 Items by Revenue')
plt.xlabel('Item')
plt.ylabel('Revenue')
plt.show()
```

```python
# Load the sleepstudy dataset from pydataset and read it's documentation.
# Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant 
# line showing the average change in reaction time.


# Load the sleepstudy dataset from pydataset
sleepstudy = data('sleepstudy')

# Create a line chart of all the individual subject's reaction times
sns.lineplot(data=sleepstudy, x='Days', y='Reaction', hue='Subject', alpha=0.5)

# Create a line chart of the average change in reaction time
sns.lineplot(data=sleepstudy, x='Days', y='Reaction', ci='sd', linewidth=2.5)

# Add a title, x-axis label, and y-axis label to the plot
plt.title('Reaction Time by Day')
plt.xlabel('Days')
plt.ylabel('Reaction Time')

# Show the plot
plt.show()
```
